```
import numpy as np
from pyspark.sql.functions import pandas_udf
from pyspark.sql.functions import PandasUDFType
import pyspark.sql.types as T
import pyspark.sql.functions as F
```

```
@pandas_udf(T.DoubleType(), functionType=PandasUDFType.GROUPED_AGG)
def distinct_avg_udf(keys, values):
  return np.mean([x[1] for x in set(zip(keys, values))])
```
