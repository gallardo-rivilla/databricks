## distinct_avg_udf

```
df_input_distinct_avg_udf = (spark
                            .createDataFrame([(1, "A", 1, "0000", 5), 
                                              (1, "A", 1, "0001", 6), 
                                              (1, "A", 1, "0002", 4), 
                                              (1, "A", 2, "0000", 5), 
                                              (1, "B", 3, "0003", 7), 
                                              (1, "B", 3, "0000", 5), 
                                              (1, "B", 4, "0003", 7)], 
                                             ["terri", "pro", "mediador", "cliente", "nota"]))

nota_media_1 = np.mean([5, 6, 4, 7])
nota_media_a = np.mean([5, 6, 4])
nota_media_b = np.mean([7, 5])
```

```
df_test_distinct_avg_udf = (df_input_distinct_avg_udf
                            .rollup("terri", "pro")
                            .agg(distinct_avg_udf("cliente", "nota").alias("nota_media")))

df_test_distinct_avg_udf.cache()
```

```
nota_media_1_test = (df_test_distinct_avg_udf
                     .filter((F.col("terri") == 1) & (F.col("pro").isNull()))
                     .collect()[0]["nota_media"])

nota_media_a_test = (df_test_distinct_avg_udf
                     .filter((F.col("terri") == 1) & (F.col("pro") == "A"))
                     .collect()[0]["nota_media"])

nota_media_b_test = (df_test_distinct_avg_udf
                     .filter((F.col("terri") == 1) & (F.col("pro") == "B"))
                     .collect()[0]["nota_media"])

assert nota_media_1 == nota_media_1_test
assert nota_media_a == nota_media_a_test
assert nota_media_b == nota_media_b_test
```
